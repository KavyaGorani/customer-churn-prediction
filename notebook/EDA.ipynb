```python
sns.countplot(x='gender', data = df)
plt.title("Gender")
```

## Most customers in the dataset are male; however, the gender distribution is relatively balanced overall.

```python
sns.countplot(x='Contract',data = df,color='orange')
plt.title("Tenure")
```

## Most customers are on month-to-month contracts, which may contribute to higher churn compared to long-term plans.

```python
sns.histplot(x='MonthlyCharges',data = df,color='green',kde=True)
plt.title("MonthlyCharges($)")
```

## Most customers have monthly charges around $20, indicating that lower-priced plans are more common in the dataset.

```python
sns.countplot(x='InternetService', data = df,color="cyan")
plt.title("InternetServices")
```

## Fiber optic internet service is the most commonly used among customers, indicating a preference for high-speed connectivity.

```python
sns.countplot(x='PaymentMethod', data = df,color="lightblue")
plt.title("PaymentMethod")
plt.xticks(rotation=45, ha="right")
```

## Customers most frequently use electronic check as their payment method, whereas automatic bank transfer and credit card payments are comparatively less common.

```python
sns.countplot(x='PaperlessBilling', data = df,color="grey")
plt.title("PaperlessBilling")
```

## Most customers have opted for paperless billing, indicating a preference for digital billing methods.

# EDA Summary:
## The exploratory analysis shows that customer churn is strongly associated with contract type, tenure, monthly charges, and internet service. Customers on month-to-month contracts, with shorter tenure and higher monthly charges, exhibit higher churn rates. Electronic check is the most common payment method, and paperless billing is widely used among customers.
